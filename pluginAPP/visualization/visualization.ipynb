```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('/Users/haotianli/Library/CloudStorage/OneDrive-Personal/MouseData/condensateID/Prox1/tables/results.csv')

data['Radius'] = np.sqrt(data['Area'] / np.pi)

plt.figure(figsize=(8, 6))

# 使用 Seaborn 的 KDE 绘制拟合曲线，并叠加直方图
sns.histplot(data['Radius'], bins=50, color='lightcoral', edgecolor='black', kde=True)

# 添加标题和标签
plt.title('Histogram and Fitted Distribution of Condensate Radius')
plt.xlabel('Radius')
plt.ylabel('Frequency')
plt.grid(True)

# 显示图像
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 读取第一个 CSV 文件的数据
data1 = pd.read_csv('/Users/haotianli/Library/CloudStorage/OneDrive-Personal/MouseData/condensateID/Prox1/tables/results_GFP.csv')
data1['Radius'] = np.sqrt(data1['Area'] / np.pi)

# 读取第二个 CSV 文件的数据
data2 = pd.read_csv('/Users/haotianli/Library/CloudStorage/OneDrive-Personal/MouseData/condensateID/Prox1/tables/results_Prox1.csv')
data2['Radius'] = np.sqrt(data2['Area'] / np.pi)

plt.figure(figsize=(8, 6))

# 绘制第一个数据集的直方图及 KDE 拟合曲线
sns.histplot(data1['Radius'], bins=50, color='green', edgecolor='black', kde=True, label='GFP')

# 绘制第二个数据集的直方图及 KDE 拟合曲线
sns.histplot(data2['Radius'], bins=50, color='skyblue', edgecolor='black', kde=True, label='Prox1')

# 添加标题、标签和图例
plt.title('Histogram and Fitted Distribution of Condensate Radius (Two Datasets)')
plt.xlabel('Radius')
plt.ylabel('Frequency')
plt.legend()
plt.grid(True)

# 显示图像
plt.show()
```
