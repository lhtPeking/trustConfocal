```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('/Users/haotianli/Library/CloudStorage/OneDrive-Personal/MouseData/condensateID/Prox1/tables/results.csv')

data['Radius'] = np.sqrt(data['Area'] / np.pi)

plt.figure(figsize=(8, 6))

# 使用 Seaborn 的 KDE 绘制拟合曲线，并叠加直方图
sns.histplot(data['Radius'], bins=50, color='lightcoral', edgecolor='black', kde=True)

# 添加标题和标签
plt.title('Histogram and Fitted Distribution of Condensate Radius')
plt.xlabel('Radius')
plt.ylabel('Frequency')
plt.grid(True)

# 显示图像
plt.show()
```
