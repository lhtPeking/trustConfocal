```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import ks_2samp

# 读取第一个 CSV 文件的数据
data1 = pd.read_csv('/Users/haotianli/Library/CloudStorage/OneDrive-Personal/MouseData/condensateID/Prox1/tables/results_GFP.csv')
data1['Radius'] = np.sqrt(data1['Area'] / np.pi)

# 读取第二个 CSV 文件的数据
data2 = pd.read_csv('/Users/haotianli/Library/CloudStorage/OneDrive-Personal/MouseData/condensateID/Prox1/tables/results_Prox1.csv')
data2['Radius'] = np.sqrt(data2['Area'] / np.pi)

# 进行双样本 K-S 检验
ks_stat, p_value = ks_2samp(data1['Radius'], data2['Radius'])

# 生成图像并在图上标注 p 值
plt.figure(figsize=(8, 6))

# 绘制第一个数据集的直方图及 KDE 拟合曲线
sns.histplot(data1['Radius'], bins=50, color='lightcoral', edgecolor='black', kde=True, label='GFP')

# 绘制第二个数据集的直方图及 KDE 拟合曲线
sns.histplot(data2['Radius'], bins=50, color='skyblue', edgecolor='black', kde=True, label='Prox1')

# 添加标题、标签和图例
plt.title('Histogram and Fitted Distribution of Condensate Radius (Two Datasets)')
plt.xlabel('Radius')
plt.ylabel('Frequency')
plt.legend()

# 标注 p 值在图上
plt.text(0.95, 0.95, f'K-S Test p-value: {p_value:.4f}', horizontalalignment='right', verticalalignment='top', 
         transform=plt.gca().transAxes, fontsize=12, bbox=dict(facecolor='white', alpha=0.5))
print(p_value)
# 显示网格和图像
plt.grid(True)
plt.show()
```
